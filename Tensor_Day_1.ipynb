```python
# Normal Distribution
nTensor = tf.random.normal(
    [3,3,3],
    mean=0.0,
    stddev=1.0,
    dtype=tf.dtypes.float32,
    seed=5,
    name=None
)
print(nTensor)
```

```python
# Uniform Distribution
uTensor = tf.random.uniform(
    [23, 23],
    low=0.0,
    high=1.0,
    validate_args=False,
    allow_nan_stats=True,
    name='Uniform'
)
print(uTensor)
```

```python
# Indexing 0-dimensional Tensor
tensor0d = tf.constant(23)
print(tensor0d)
#print(tensor0d[0])
```

```python
# Indexing 1-dimensional Tensor
print(tensor1d[3])
print(tensor1d[:])
print(tensor1d[0:3])
print(tensor1d[0:])
print(tensor1d[:3])
print(tensor1d[0:-1])
print(tensor1d[:-1])
print(tensor1d[3:-1])
print(tensor1d[0:4:2])
print(tensor1d[0:4:-1])
```

```python
# Indexing 2-dimensional Tensor
print(tensor2d[0:4,0:2])
print(tensor2d[0:,:2])
print(tensor2d[0:3,0:])
print(tensor2d[:,:])
print(tensor2d[:4,:])
print(tensor2d[:,0:2])
print(tensor2d[0,0:2])
print(tensor2d[0:4,])
```

```python
# Indexing 3-dimensional Tensor
print(tensor3d[0:4,0:2,0:3])
print(tensor3d[0:,:2,:])
print(tensor3d[0:3,0:,0:2])
print(tensor3d[:,:,:])
print(tensor3d[:4,:,0:1])
print(tensor3d[:,0:2,:2])
print(tensor3d[0,0:2,1:])
print(tensor3d[0:4,2,1])
```

```python
# Indexing 4-dimensional Tensor
print(tensor4d[0:4,0:2,0:3,:])
print(tensor4d[0:,:2,:,0:2])
print(tensor4d[0:3,0:,0:2,0:])
print(tensor4d[:,:,:,:])
print(tensor4d[:4,:,0:1,0:1])
print(tensor4d[:,0:2,:2,2])
print(tensor4d[0,0:2,1:,2:])
print(tensor4d[0:4,2,1,:2])
```

```python
# Indexing 5-dimensional Tensor
print(tensor5d[0:4,0:2,0:3,:,0:2])
print(tensor5d[0:,:2,:,0:2,0:3])
print(tensor5d[0:3,0:,0:2,0:,:3])
print(tensor5d[:,:,:,:,:])
print(tensor5d[:4,:,0:1,0:1,:])
print(tensor5d[:,0:2,:2,2,2])
print(tensor5d[0,0:2,1:,2:,2:])
print(tensor5d[0:4,2,1,:2,1:])
```
